```python
OpenFile = open('students.json')
#JsonData = json.load(OpenFile)

Student_Data = [json.loads(line) for line in OpenFile]
```

```python
client = client = pymongo.MongoClient("mongodb+srv://student:1234@database.bkozt3b.mongodb.net/?retryWrites=true&w=majority")
db = client.test
records = db.students_database
```

```python
for Student in Student_Data:
  records.insert_one(Student)
```

```python
x = records.find({})
file = {'_id':[],'name':[],'quiz':[],'homework':[],'exam':[]}
stud = pd.DataFrame(file)
for i in x:
    file['_id'].append(i['_id'])
    file['name'].append(i['name'])
    Scores = [s['score'] for s in i['scores']]
    file['quiz'].append( Scores[0])
    file['homework'].append( Scores[1])
    file['exam'].append( Scores[2])
stud = pd.DataFrame(file)
display(stud)
```

 Student who scored maximum score

```python
total = stud.quiz + stud.homework + stud.exam
total.max()
stud['Total'] = total
Max_score = stud[stud['Total'] == total.max()]
display(Max_score)
```
